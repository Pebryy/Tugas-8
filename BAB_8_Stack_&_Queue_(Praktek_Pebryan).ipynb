```python
## Antrian Pelanggan (Queue)
from collections import deque

antrian = deque()

n = int(input("Masukkan jumlah pelanggan: "))
```

```python
for i in range(n):
  nama = input(F"Nama pelanggan ke-{i+1}: ")
  antrian.append(nama)  #enqueque
```

```python
if antrian:
  dilayani = antrian.popleft() #dequeque
  print("\nPelanggan yang dilayani:", dilayani)
```

```python
print("\nSisa Antrean:")
for pelanggan in antrian:
  print(pelanggan)
```

```python
## Pemeriksa Keseimbangan Kurung
def cek_kurung(ekspresi):
  stack = []

  pasangan = {
      ')': '(',
      '}': '{',
      ']': '['
  }

  for karakter in ekspresi:

    if karakter in "([{":
      stack.append(karakter)

    elif karakter in ")]}":

      if len(stack) == 0:
        return False

      atas = stack.pop()

      if atas != pasangan[karakter]:
        return False

  return len(stack) == 0


ekspresi = input("Masukkan ekspresi matematika: ")

if cek_kurung(ekspresi):
  print("Ekspresi valid")
else:
  print("Ekspresi tidak valid")
```

```python
## Simulasi Antrian Rumah Sakit Prioritas
from collections import deque

antrian_darurat = deque()
antrian_umum = deque()

n = int(input("Jumlah pasien: "))

for i in range(n):

    data = input(
        "Masukkan Nama dan Jenis Pasien (darurat/umum): "
    ).split()

    nama = data[0]

    # hilangkan kurung dan ubah jadi huruf kecil
    jenis = data[1].replace("(", "").replace(")", "").lower()

    if jenis == "darurat":
        antrian_darurat.append(nama)
    else:
        antrian_umum.append(nama)

print("\nUrutan Pelayanan:")

while antrian_darurat or antrian_umum:
    if antrian_darurat:
        print(antrian_darurat.popleft())

    if antrian_umum:
        print(antrian_umum.popleft())
```

```python
## (Proyek) – Sistem Manajemen Browser History dan Download Manager
from collections import deque


# Struktur Data
history = []
download_queue = deque()

total_kunjungan = 0
download_selesai = 0


# Fungsi Browser History
def kunjungi_halaman():
    global total_kunjungan

    url = input("Masukkan URL: ")

    history.append(url)

    total_kunjungan += 1

    print("Halaman berhasil dikunjungi.")


def back():
    if len(history) == 0:
        print("History kosong.")
    else:
        halaman = history.pop()
        print("Kembali dari:", halaman)


def tampil_history():
    if len(history) == 0:
        print("History kosong.")
        return

    print("\n===== HISTORY =====")

    for halaman in reversed(history):
        print(halaman)


# Fungsi Download Manager
def tambah_download():
    file = input("Nama file: ")

    download_queue.append(file)

    print("File masuk antrian download.")


def proses_download():
    global download_selesai

    if len(download_queue) == 0:
        print("Tidak ada file dalam antrian.")

    else:
        file = download_queue.popleft()

        download_selesai += 1

        print("Download selesai:", file)


def tampil_download():
    if len(download_queue) == 0:
        print("Antrian kosong.")
        return

    print("\n===== ANTRIAN DOWNLOAD =====")

    for file in download_queue:
        print(file)


# Statistik
def statistik():

    print("\n===== STATISTIK =====")

    print("Total halaman dikunjungi     :", total_kunjungan)

    print("History tersimpan            :", len(history))

    print("Download selesai            :", download_selesai)

    print("Download menunggu           :", len(download_queue))



# Menu Utama
while True:

    print("\n========================")
    print(" SISTEM STACK & QUEUE ")
    print("========================")
    print("1. Kunjungi Halaman")
    print("2. Back")
    print("3. Tampilkan History")
    print("4. Tambah Download")
    print("5. Proses Download")
    print("6. Tampilkan Queue Download")
    print("7. Statistik")
    print("0. Exit")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        kunjungi_halaman()

    elif pilihan == "2":
        back()

    elif pilihan == "3":
        tampil_history()

    elif pilihan == "4":
        tambah_download()

    elif pilihan == "5":
        proses_download()

    elif pilihan == "6":
        tampil_download()

    elif pilihan == "7":
        statistik()

    elif pilihan == "0":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid.")
```
